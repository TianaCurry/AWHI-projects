# joyplot 

```python
import pandas as pd 
import joypy
from matplotlib import pyplot as plt
from matplotlib import cm
```

```python
#for mathplotlib, old library, this command makes sure that graphs do not open outside the jupyter notebook
%matplotlib inline
```

```python
#Testing example joy plot
# Import Data
mpg = pd.read_csv("https://github.com/selva86/datasets/raw/master/mpg_ggplot2.csv")

# Draw Plot
plt.figure(figsize=(16,10), dpi= 80)
fig, axes = joypy.joyplot(mpg, column=['hwy', 'cty'], by="class", ylim='own', figsize=(14,10))

# Decoration
plt.title('Joy Plot of City and Highway Mileage by Class', fontsize=22)
plt.show()
```

```python
start_year= pd.read_csv('AWHI_researcher_start_year.csv')
start_year.head()
```

```python
#Total count of value in each column in the data set and define if non-null object orfloat64
start_year.info()
```

```python
#removing null spaces from csv start_year_complete of start_year in the column SI_START_YEAR
start_year_complete=start_year[pd.notnull(start_year['SI_START_YEAR'])]
```

```python
#Total count of value in each column in the data set and define if non-null object orfloat64
start_year_complete.info()
```

```python
#Total amount of enteries for components in SI_UNIT
start_year_complete['SI_UNIT'].value_counts()
```

```python
#checking for which rows in SI_UNIT have Castle 
start_year_complete['SI_UNIT'] != 'Castle'
```

```python
# taking out the values in for SI_UNIT->Castle

start_year_complete=start_year_complete[start_year_complete['SI_UNIT'] != 'Castle']
```

```python
start_year_complete.head()
```

# Joyplot using joypy for Smithsonian Women Researchers 

```python
plt.figure(figsize=(16,10), dpi= 80)

fig, axes = joypy.joyplot(start_year_complete, column='SI_START_YEAR', by="SI_UNIT",ylim='own',xlim=2020, figsize=(14,10), colormap=cm.Set2)

# Decoration
plt.title('RESEARCHER START YEAR', fontsize=22)
plt.show()
```
